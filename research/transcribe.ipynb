## Parallel Limit
Limit of 10 parallel jobs!

```python
started_job_names = []
finished_job_names = []
```

```python
def monitor():
    print('Progress - {:.0%} Completed - {} Running'.format(float(len(finished_job_names) / len(uploaded)), len(started_job_names)))
    for job_name in started_job_names:
        status = transcribe.get_transcription_job(TranscriptionJobName=job_name)
        status_pretty = status['TranscriptionJob']['TranscriptionJobStatus']
        #print(job_name, " - ", status_pretty)
        if status_pretty in ['COMPLETED', 'FAILED']:
            finished_job_names.append(job_name)
            started_job_names.remove(job_name)
```

```python
# Start a new transcription job for every mp3 set
s3_base = 'https://s3.amazonaws.com/cbc-ds-podcasts/'
started_job_names = []
finished_job_names = []
for mp3_name in uploaded:
    
    # Check for number of parallel jobs, max 10
    while len(started_job_names) >= 8:
        monitor()
        time.sleep(5)
    
    job_name = mp3_name
    job_uri = s3_base + mp3_name
    
    print('Starting...', job_name)
    
    try:
        status = transcribe.get_transcription_job(TranscriptionJobName=job_name)
        if status['TranscriptionJob']['TranscriptionJobStatus'] in ['Completed', 'FAILED']:
            finished_job_names.append(job_name)
        else:
            started_job_names.append(job_name)
        print('! Already found... skipping creation.')
    except:
        transcribe.start_transcription_job(
            TranscriptionJobName=job_name,
            Media={'MediaFileUri': job_uri},
            MediaFormat='mp3',
            LanguageCode='en-US'
        )
        started_job_names.append(job_name)
```

```python
def upload_to_s3(bucket, local_filename, s3_filename):
    s3_resource.Object(bucket, s3_filename).upload_file(
        Filename=local_filename)

def output_transcript_s3(job_names):
    output = []
    
    for name in job_names:
        print('Uploading ', name)
        status = transcribe.get_transcription_job(TranscriptionJobName=job_name)
        status_pretty = status['TranscriptionJob']['TranscriptionJobStatus']
        if status_pretty not in ['COMPLETED']:
            print('not completed, skipping...')
            continue
        transcript_uri = status['TranscriptionJob']['Transcript']['TranscriptFileUri']
        response = requests.get(transcript_uri)
        json = response.json()
        transcript = json['results']['transcripts'][0]['transcript']
        s3_name = name+'.txt'
        tmpname = 'temp.txt'
        with open(tmpname, 'wb') as f:
            f.write(transcript)
        # upload fname to s3
        upload_to_s3(bucket, tmpname, s3_name)
        output.append(s3_name)
    
    return output
```
