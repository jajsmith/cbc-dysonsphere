```python
from scipy.sparse import csr_matrix
from pandas.api.types import CategoricalDtype

sessionId_c = CategoricalDtype(sorted(read_data.sessionId.unique()), ordered=True)
contentId_c = CategoricalDtype(sorted(read_data.contentId.unique()), ordered=True)

row = read_data.sessionId.astype(sessionId_c).cat.codes
col = read_data.contentId.astype(contentId_c).cat.codes
sparse_matrix = csr_matrix((read_data["eventId"], (row, col)), \
                           shape=(sessionId_c.categories.size, contentId_c.categories.size))
```
